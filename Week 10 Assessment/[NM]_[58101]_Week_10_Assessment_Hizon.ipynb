## **Week 10 Assessment**

Estimate the root of the function f(x) = e^-x -x in four decimal pplaces with interval [0, 1], and with tolerance limit of 0.0002

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
def function(x):
  return np.exp(-x) - x

xList = np.linspace(0, 1)
```

```python
plt.figure(num=1, dpi=120)
plt.plot(xList, function(xList), label = 'Given Function', color = 'cyan')
plt.plot(xList, xList*0, "--", label='y = 0', color ='pink')

plt.legend()
```

```python
root = fsolve(function, 0.0002)
root
```
