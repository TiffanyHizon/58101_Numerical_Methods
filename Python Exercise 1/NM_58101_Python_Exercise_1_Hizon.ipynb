## **Python Exercise 1**

Create a program using Python that allows a student to type his/her name, the name of the course, and that computes his/her semestral grade

$$ SemestralGrade = 30\%\ * PG + 30\%\ * MG + 40\%\ x FG $$

```python
print("Hi Klasmeyt! Welcome to Grade Calculator!")

def grade():
  Name = input("Enter Name: ")
  Program = input("Enter Program: ")
  Course = input("Enter a Course: ")

  prelimGrade = float(input("Enter Prelim Grades: "))
  midtermGrade = float(input("Enter Midterm Grades: "))
  finalGrade = float(input("Enter Final Grades: "))

  semestralGrade = ((prelimGrade * 0.3) + (midtermGrade * 0.3) + (finalGrade * 0.4))
  #semestralGrade = round(semestralGrade, 2)
  print("Hi Klasmeyt, ", Name, "from ", Program, "your semestral grade in ", Course,"is: {0:.2f}".format(semestralGrade))
  
  if semestralGrade > 70.00:
    print("Congratulations, Klasmeyt!")
  else:
    print("You did well, Klasmeyt!")
    
grade()
```
