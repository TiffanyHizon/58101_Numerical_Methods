## **Matrix and its Operations**

Create a NumPy program that performs the operations of the given matrices:
1. Sum = A + B
2. Difference1 = B - A
3. Difference2 = A - B

$$A = \begin{bmatrix} -5&0\\4&1 \\\end{bmatrix} , B = \begin{bmatrix} 6 & 3 \\ 2 & 3 \end{bmatrix} $$

```python
import numpy as np
```

```python
A = np.array([
              [-5, 0],
              [4, 1]
              ])
B = np.array([
              [6, -3],
              [2, 3]
              ])
```

**Addition**

```python
#Eager Execution

Sum = A + B
Sum
```

```python
#Functional Method

Sum = np.add(A, B)
Sum
```

**Subtraction - Difference 1**

```python
#Eager Execution

Difference1 = B - A
Difference1
```

```python
#Functional Method

Difference1 = np.subtract(B, A)
Difference1
```

**Subtraction - Difference 2**


```python
#Eager Execution

Difference2 = A - B
Difference2
```

```python
#Functional Method

Difference2 = np.subtract(A, B)
Difference2
```
